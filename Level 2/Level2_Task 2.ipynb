```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load our dataset
df = pd.read_csv("Dataset .csv")
```

```python
df.head(5)
```

```python
#most common combinations of cuisines in the dataset
top_comb_cuisines=df['Cuisines'].value_counts().head()
print(top_comb_cuisines)
```

```python
plt.figure(figsize=(6, 4))
top_comb_cuisines.plot(kind='bar')
plt.title('Top combination of Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Counts')
plt.show()
```

```python
#calculate cuisine combinations tend to have higher ratings
avg_rating_cuisine_combination=df.groupby('Cuisines')['Aggregate rating'].mean().nlargest(20)
```

```python
print("cuisines combinations haveing higher ratings: ")
print(avg_rating_cuisine_combination)
```

```python
plt.figure(figsize=(6, 4))
avg_rating_cuisine_combination.plot(kind='bar')
plt.title('Average ratings of top combination of Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Aggregate rating')
plt.show()
```
