```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load our dataset
df = pd.read_csv("Dataset .csv")
```

```python
df.head(5)
```

```python
#percentage of restaurants that offer online delivery
online_delivery_restaurants_counts=df['Has Online delivery'].value_counts().get('Yes',0)
total_restaurnts=len(df)
percentage = (online_delivery_restaurants_counts/total_restaurnts)*100
print("percentage of restaurants that offer online delivery:" )
print(percentage)
```

```python
# Compare the average ratings of restaurants with and without online delivery
#average_rating_with_online_delivery = 
rating_online_delivery= df[df['Has Online delivery'] == "Yes"]['Aggregate rating']
rating_no_online_delivery = df[df['Has Online delivery'] == "No"]['Aggregate rating']

avg_rating_online_delivery= rating_online_delivery.mean()
avg_rating_no_online_delivery =rating_no_online_delivery.mean()
print(f'Average rating of restaurants with online delivery: {avg_rating_online_delivery}')
print(f'Average rating of restaurants without online delivery: {avg_rating_no_online_delivery}')
```

```python
#visulaization of our findings
plt.bar(['With Online Delivery', 'Without Online Delivery'], [avg_rating_online_delivery,avg_rating_no_online_delivery],color=['green','orange'])
plt.xlabel('Status of online delivery')
plt.ylabel('Average Rating')
plt.title('Comparison of Average Ratings by Online Delivery')
plt.show()
```
