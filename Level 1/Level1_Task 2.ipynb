```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load our dataset
df = pd.read_csv("Dataset .csv")
```

```python
df.head(5)
```

```python
# count the cities in data
city_counts= df['City'].value_counts()
```

```python
# find the city with highest number of restaurants
print("City with the highest number of restaurants:")
print(city_counts.index[0], "with",city_counts.values[0], "restaurants")
```

```python
# Average rating for restaurants in each city
average_rating_of_restaurants_by_city = df.groupby('City')['Aggregate rating'].mean()
print("Average ratings of restaurants bny city:")
print(average_rating_of_restaurants_by_city)
```

```python
#city with the highest average rating
highest_rating_city = average_rating_of_restaurants_by_city.idxmax()
print("City with the highest average rating:")
print(highest_rating_city)
```

```python
#lets have some visualization for findings

#visulaization with Bar chart of cities with highest number of restaurants
top_cities =city_counts.nlargest(6)
plt.figure(figsize=(8,6))
plt.bar(top_cities.index, top_cities.values, color="pink", width=0.5)
plt.xlabel('Some top cities')
plt.ylabel('Number of Restaurants')
plt.title('Top Cities by Number of Restaurants')
plt.show()
```

```python
#visulaization with scatter plot of average ratings for some of the cities
plt.scatter(average_rating_of_restaurants_by_city.index[0:25], average_rating_of_restaurants_by_city.values[0:25],alpha=0.6, edgecolors='b', linewidths=2)
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Each City')
plt.xticks(rotation=90)
plt.show()
```

```python
#visulaization for average highest ratings city
highest_rating_city = average_rating_of_restaurants_by_city.nlargest(3)

plt.barh(highest_rating_city.index,highest_rating_city.values,color="skyblue")
plt.xlabel('Average Rating')
plt.ylabel('City')
plt.title('Top Cities by Average Rating')
plt.show()
```
