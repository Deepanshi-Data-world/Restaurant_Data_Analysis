```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load our dataset
df = pd.read_csv("Dataset .csv")
```

```python
df.head(5)
```

```python
#calculate of counts of price ranges among the restaurants
price_range_counts = df['Price range'].value_counts()
```

```python
#sort the values 
p_r_c= price_range_counts.sort_index()
```

```python
#create a bar chart to visualize
plt.figure(figsize=(6, 4))
plt.bar(p_r_c.index,p_r_c.values,color=['green', 'blue', 'yellow','pink'])
plt.title('price ranges distribution')
plt.xlabel('price range')
plt.ylabel('Counts')
plt.show()
```

```python
#percentage of restaurants in each price range category
total_restaurants=len(df)
percentage = (p_r_c/total_restaurants)*100
print("percentage of restaurants in each price range category: ")
print(percentage)
```
